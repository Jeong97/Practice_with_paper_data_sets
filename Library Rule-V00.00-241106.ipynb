```python
# metrics.py로 저장해서 호출하기
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

class PerformanceMetrics:
    def __init__(self, y_true, y_pred):
        self.y_true = y_true
        self.y_pred = y_pred

    def mean_absolute_error(self):
        return mean_absolute_error(self.y_true, self.y_pred)

    def mean_squared_error(self):
        return mean_squared_error(self.y_true, self.y_pred)

    def root_mean_squared_error(self):
        return np.sqrt(self.mean_squared_error())

    def r2_score(self):
        return r2_score(self.y_true, self.y_pred)

    def summary(self):
        return {
            'MAE': self.mean_absolute_error(),
            'MSE': self.mean_squared_error(),
            'RMSE': self.root_mean_squared_error(),
            'R2 Score': self.r2_score()
        }
```

```python
# Visualization.py
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 산점도 시각화
def plot_scatter(x, y, title="Scatter Plot", xlabel="X-axis", ylabel="Y-axis", legend_label="Data Points"):
    plt.figure(figsize=(8, 6))
    plt.scatter(x, y, color='blue', edgecolor='black', facecolors='none', s=50, label=legend_label)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()

# 성능 비교 시각화
def plot_performance(y_true, y_pred, title="Model Performance Comparison", xlabel="Actual Values", ylabel="Predicted Values"):
    plt.figure(figsize=(8, 6))
    plt.plot(y_true, y_pred, 'o', markerfacecolor='none', label="Predicted vs Actual")
    plt.plot(y_true, y_true, 'r--', label="Ideal Prediction (y=x)")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()

# 분포 시각화
def plot_distribution(data, title="Data Distribution", xlabel="Value", ylabel="Frequency", color="blue"):
    plt.figure(figsize=(8, 6))
    sns.histplot(data, kde=True, color=color)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()

# 상관관계 시각화
def plot_correlation(x, y, correlations, title="Correlation Plot", xlabel="Frequency (Hz)", ylabel="Correlation", labels=None):
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, 'o-', color='blue', label=labels[0] if labels else "Correlation with Real Z")
    plt.plot(x, correlations, 's-', color='orange', label=labels[1] if labels else "Correlation with Imaginary Z")
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.show()

# Nyquist Plot
def plot_nyquist(real, imag, title="Nyquist Plot", xlabel="Real Z (Ω)", ylabel="Imaginary Z (Ω)"):
    plt.figure(figsize=(8, 6))
    plt.plot(real, -imag, 'o', color='blue', markerfacecolor='none', label="Nyquist Data")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(True)
    plt.show()

# Bode Plot (Log Scale)
def plot_bode(frequency, magnitude, phase, title="Bode Plot", mag_label="Magnitude (dB)", phase_label="Phase (degrees)"):
    fig, ax1 = plt.subplots(2, 1, figsize=(10, 8))

    # Magnitude plot
    ax1[0].semilogx(frequency, magnitude, 'b-', label="Magnitude")
    ax1[0].set_title(f"{title} - Magnitude")
    ax1[0].set_xlabel("Frequency (Hz)")
    ax1[0].set_ylabel(mag_label)
    ax1[0].grid(True, which='both', linestyle='--', linewidth=0.5)

    # Phase plot
    ax1[1].semilogx(frequency, phase, 'g-', label="Phase")
    ax1[1].set_title(f"{title} - Phase")
    ax1[1].set_xlabel("Frequency (Hz)")
    ax1[1].set_ylabel(phase_label)
    ax1[1].grid(True, which='both', linestyle='--', linewidth=0.5)

    plt.tight_layout()
    plt.show()
```

```python
# main.py
import numpy as np
from Visualization import plot_nyquist, plot_bode

# 예제 데이터 생성
frequency = np.logspace(1, 4, num=50)  # 10Hz ~ 10kHz 사이의 주파수
magnitude = 20 * np.log10(1 / (1 + (frequency / 1000)**2))  # 임의의 진폭(dB)
phase = -np.arctan(frequency / 1000) * (180 / np.pi)  # 임의의 위상(degrees)
real = np.cos(phase * np.pi / 180) * magnitude  # 임의의 실수부
imag = np.sin(phase * np.pi / 180) * magnitude  # 임의의 허수부

# Nyquist Plot 호출
plot_nyquist(real, imag, title="Example Nyquist Plot")

# Bode Plot 호출
plot_bode(frequency, magnitude, phase, title="Example Bode Plot")
```
